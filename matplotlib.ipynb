#### <p style='text-align:center'> Visualizatoin using Matplotlib Library </p> ####



Matplotlib has two intrfaces MATLAB-style state based interface and a more powerful object-oriented interface. The first is simpler, while the second is more suitable for complicated situations. 

##### MATLAB-style Interface #####

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0, 10, 100)
```

```python
plt.figure()  # create a plot figure
```

```python
plt.subplot(2,1,1)  #(rows, columns, panel number)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2) # create the second panel and set current axis 
plt.plot(x, np.cos(x))
```

##### Object-oriented interface ##### 

```python
fig, ax = plt.subplots(2)   # create a grid of plots, ax is an array of two axes objects 
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

to add legend: ax.legend 

Two ways to add labels, title, and change limits in object oriented interface: 

**First:** 

ax.set_xlabel
ax.set_ylabel
ax.set_xlim
ax.set_ylim
ax.set_title


**Second:**

ax.set( xlim=(0,10), ylim=(-2,2), xlabel='x' , ylabel='sin(x)' , title='A simple Plot' ) 

***

### Line Plot ##

```python
fig = plt.figure()
ax= plt.axes()

x = np.linspace(0, 10, 10000)
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cox(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

#### Colors and markers ####

color codes: 
'b': blue  
'k': black  
'w': white  
'g': green  
'y': yellow  

linestyle:  
'solid' or '-'  
'dashed' or '--'  
'dashdot' or '-.'  
'dotted' or ':'   

We can also do them together '-g' for solid green, or ':b' for dotted blue line

```python
plt.plot(x, np.sin(x-0), color='blue', linestyle='solid')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), ':r')
plt.plot(x, np.sin(x-3), '--y')
```

#### Adjusting the plot: Axes Limits #### 

```python
plt.plot(x, np.sin(x), 'y--')

plt.xlim(-1, 11)
plt.ylim(-2, 2)
```

```python
#alternatively we can use plt.axis (not axes) using keywords like 'tight', 'on', 'off', 'square', 'image'
plt.plot(x, np.sin(x))
plt.axis('tight');
```

#### Labeling Plots #### 

```python
plt.plot(x, np.sin(x), 'g-', label='sin(x)')
plt.plot(x, np.cos(x), '--y', label='cos(x)')
plt.title('A sine Cure')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.legend()
```

### Scatter Plot ### 

Very similar to plt.plot, but plt.scatter gives more control over individual point properties (size, face color, edge color, etc). The cost is more expensive computation so when we have more than ~1000 points to plot, plt.plot is more effecient. 

```python
x = np.linspace(0,10, 30)
y = np.sin(x)

plt.plot(x, y, 'ok') # o for circle and k for black
```
